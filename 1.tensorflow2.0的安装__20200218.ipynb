# 2.制作gpu版本的tensorflow2.0虚拟环境


查看conda的环境

conda env list





创建环境
conda create -n tensorflow_gpu python=3.7


激活环境
conda activate tensorflow_gpu



# 3.安装指定版本的tensorflow

pip install -i https://pypi.tuna.tsinghua.edu.cn/simple tensorflow-gpu==2.0.0
